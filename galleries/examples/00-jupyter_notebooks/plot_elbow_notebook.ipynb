### An example showing the plot_elbow method used by a scikit-learn classifier

In this example, we'll be plotting a `elbow` to describe the classifier model performance
using the **breast_cancer**, **iris** or **digits** dataset from scikit-learn.
- `KMeans`

Here, we'll be using the `scikitplot.cluster.plot_elbow` method.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris as data_3_classes
from sklearn.model_selection import train_test_split

import numpy as np
np.random.seed(0)
# importing pylab or pyplot
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 5)

# Import scikit-plot
import scikitplot as sp

sp.__version__
```

```python
sp.estimators.plot_elbow?
```

```python
# Load the data
X, y = data_3_classes(return_X_y=True, as_frame=False)
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.5, random_state=0
)
```

```python
# Create an instance of the LogisticRegression
model = KMeans(n_clusters=4, random_state=1)

# Plot!
ax = sp.estimators.plot_elbow(
    model, X_train, cluster_ranges=range(1, 11)
)
```
